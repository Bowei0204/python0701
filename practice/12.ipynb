```python
import pandas as pd

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
df = pd.read_csv(url)

# 移除重複的國家名稱，只保留第一個出現的
df_unique_countries = df.drop_duplicates(subset=['country'])

# 儲存為新的 CSV 檔案
df_unique_countries.to_csv("各國 GDP 指數_去重.csv", index=False)

print("檔案已成功儲存為各國 GDP 指數_去重.csv")
```

```python
import os
import csv
import requests

url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

file_path = os.path.abspath(__name__) #檔案的絕對位置
current_dir_name = os.path.dirname(file_path) #取得目前的目錄
file_position = os.path.join(current_dir_name,'data','各國的GDP.csv') #操作檔案的絕對位置

response = requests.get(url)
data = response.text



# 寫入CSV文件
with open(file_position, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, ['country', 'year', 'pop', 'continent', 'lifeExp', 'gdpPercap'])
    writer.writeheader()
    writer.writerows(data)

print(f"完成存檔 {file_position}")
```
