```python
import matplotlib.pyplot as plt

student1 = {'chinese':78,
           'english':80,
           'math':92,
           'histroy':75,
           'discover':85}

student2 = {'chinese':70,
           'english':90,
           'math':100,
           'histroy':63,
           'discover':77}

subject = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot()

axes.plot(subject,score1,'bo-',label='student1')
axes.plot(subject,score2,'ro-',label='student2')

axes.set_title('Table Of Students')
axes.set_xlabel('Subject')
axes.set_ylabel('Scores')

x_location = axes.get_xticks()
for i in x_location:
    axes.annotate(str(score1[i]),xy=(i-0.1,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i-0.1,score2[i]+0.5))
axes.grid(axis='y')
axes.legend()
plt.show()
```
