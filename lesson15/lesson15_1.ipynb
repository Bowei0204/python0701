```python
import matplotlib.pyplot as plt

student1 = {'chinese':78,
           'english':80,
           'math':92,
           'histroy':75,
           'discover':85}

student2 = {'chinese':70,
           'english':90,
           'math':100,
           'histroy':63,
           'discover':77}

subject = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot()

axes.plot(subject,score1,'bo-',label='student1')
axes.plot(subject,score2,'ro-',label='student2')

axes.set_title('Table Of Students')
axes.set_xlabel('Subject')
axes.set_ylabel('Scores')

x_location = axes.get_xticks()
for i in x_location:
    axes.annotate(str(score1[i]),xy=(i-0.1,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i-0.1,score2[i]+0.5))
axes.grid(axis='y')
axes.legend()
plt.show()
```

```python
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)
#font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8)
#使用codespace 要安裝中文字型
#install 要加-y
#命令列:sudo apt-get update -y
#命令列:sudo apt-get install -y fonts-wqy-zenhei
#說明網址:https://askubuntu.com/questions/490829/how-can-i-install-chinese-fonts-on-kubuntu-14-04
#字型位置:/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc

import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=10)

student1 = {'國文':78,
           '英文':80,
           '數學':92,
           '歷史':75,
           '地理':85}

student2 = {'國文':70,
           '英文':90,
           '數學':100,
           '歷史':63,
           '地理':77}

subject = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot()

axes.plot(subject,score1,'bo-',label='學生1')
axes.plot(subject,score2,'ro-',label='學生2')

axes.set_title('分數表',fontproperties = font)
axes.set_xlabel('科目',fontproperties = font)
axes.set_ylabel('分數',fontproperties = font)
axes.set_xticklabels(subject,fontproperties = font)

x_location = axes.get_xticks()
for i in x_location:
    axes.annotate(str(score1[i]),xy=(i-0.1,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i-0.1,score2[i]+0.5))
axes.grid(axis='y')
axes.legend(prop = font)
plt.show()
```
