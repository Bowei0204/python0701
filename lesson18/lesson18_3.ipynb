```python
import os
current_path = os.path.dirname(os.path.abspath(__name__))
folder_path = os.path.join(current_path,'每日各站進出站人數')
fileName_list = []
for filename in os.listdir(folder_path):
    if '每日各站進出站人數' in filename:
        fileName_list.append(os.path.join(folder_path,filename))
```

```python
import json
import pandas as pd
from pprint import pprint
pd.read_csv(fileName_list[0])

station_path = os.path.join(folder_path,"車站基本資料集.json")
with open(station_path,encoding="utf-8") as file:
    station_content = json.load(file)

station_df = pd.DataFrame(station_content)
station_df['stationCode'] = station_df["stationCode"].astype(int)

station_df.info()
```

```python
inOut_df = pd.read_csv(fileName_list[0])
inOut_df.info()
```

```python
df1 = pd.merge(inOut_df,station_df,left_on='staCode',right_on='stationCode')
df2 = df1.reindex(columns=['trnOpDate','stationName','gateInComingCnt','gateOutGoingCnt'])
df3 = df2.rename(columns={'trnOpDate':'日期',
                          'stationName':'站名',
                          'gateInComingCnt':'進站人數',
                          'gateOutGoingCnt':'出站人數'
                          })
df3
```

```python
df3['日期'] = pd.to_datetime(df3['日期'].astype(str))
```
