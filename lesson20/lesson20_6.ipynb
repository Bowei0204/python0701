```python
from tools import get_path,station_info,merge_dataFrame
import pandas as pd

fileName_list = get_path()
stations_df = station_info()
all_in_out = merge_dataFrame(fileName_list, stations_df)
station_in_out= pd.concat(all_in_out)
station_in_out1 = station_in_out.sort_values(by='日期')
station_in_out1
```

```python
years_month = station_in_out1['日期'].dt.strftime('%Y-%m')
station_in_out2 = station_in_out1.groupby(by=[years_month,'站名'])[['進站人數','出站人數']].aggregate([('最少','min'),('最多','max')])
```

```python
bool1 = station_in_out2['進站人數']['最少'] < 100
bool1
```

```python

station_in_out2[bool1]
```

```python
bool2 = (station_in_out2['進站人數']['最少'] < 100) & (station_in_out2['進站人數']['最多'] < 100)

station_in_out2[bool2]
```
